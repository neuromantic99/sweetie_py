```python
import sima
import sima.motion
import numpy as np
import sima.segment
from sima.ROI import ROI as roi
import matplotlib.pyplot as plt

def create_dataset(filepath):
    sequences = [sima.Sequence.create('TIFF', filepath)]
    #iterables = [[MultiPageTIFF(file)]]
    dataset = sima.ImagingDataset(sequences, None)
    return dataset
    
def channel_a(dataset):
    a = dataset[:,::2]
    return a
    
def channel_b(dataset):
    return dataset[:,1:2]

def motion_correction(filepath, data, correctname):
    bothchannels = create_dataset(filepath)
    datasetpath = data
    dataset = channel_a(bothchannels)
    correctionapproach = sima.motion.HiddenMarkov2D(max_displacement=[40, 70], verbose=False)
    mcorrected = correctionapproach.correct(dataset, datasetpath)
    mcorrected.export_frames([[[correctname]]], fmt='TIFF16')
    return mcorrected

def segment(data):
    dataset = sima.ImagingDataset.load(data)
    #iterable = tf('C:\\Users\\zcbtjro\\Pictures\\tester\\MFB_test.tif').asarray()
    segmentation_approach = sima.segment.STICA()
    dataset.segment(segmentation_approach, 'auto_ROIs')
    print 'DONE!'

def extract_signals(data, csvname):
    dataset = sima.ImagingDataset.load(data)
    #datasetpath = 'C:\\Users\\zcbtjro\\Pictures\\'
    dataset.extract(remove_overlap = False)
    dataset.export_signals(csvname, fmt = 'csv')
    #x = dataset.roi(mask=None, polygons=None, label=None, tags=None, id=None, im_shape=None)
    print 'DONE!'


    
def plot_signals(csvfile):
    with open(csvfile, 'rb') as f:
        f.readline()
        f.readline()
        f.readline()

        lista = []
        for row in f:
            lista.append(row)

    listb = []
    for i in range(0, len(lista)):
        x = lista[i].split() 
        listb.append(x)

    rawarr = np.array(listb)
    roi = []


    for i in range(2, len(rawarr[1])):
        roi.append(rawarr[:, i])

    roiarr = np.array(roi)
    t = np.arange(0, len(rawarr))
    p = t.reshape(1, len(rawarr))
    v = np.repeat(p, len(roiarr), axis = 0)

    plt.close('all')
    fig = plt.figure()
    f, axarr = plt.subplots(6, sharex = True, sharey = True)


    for i in range(0, 6):
           fig = axarr[i].plot(v[i], roiarr[i], c=np.random.rand(3,1))
           axarr[i].axes.get_yaxis().set_visible(False)
    print roiarr.shape
    plt.show()
    
filepath = 'C:\\Users\\zcbtjro\\Pictures\\MFB_running\\PN27_withrunning_002.tif'
#insert name and path of tif here

suffix = filepath[:-4]
data = filepath + '.sima'
csvname = suffix.rsplit('\\', 1)[-1] + '.csv'
correctname = suffix.rsplit('\\', 1)[-1] + '_corrected' '.tif' 
csvfile = suffix + '.csv'

#run each desired function using the arguments generated here
# e.g. run create_dataset(filepath) then motion_correction(filepath, data, correctname) etc




```
